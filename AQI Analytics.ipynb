#### **TOP AND BOTTOM POLLUTED AREAS**

```python
start_date = "2024-12-01"
end_date   = "2025-05-31"

aqi_6m = aqi[
    (aqi["date"] >= start_date) &
    (aqi["date"] <= end_date)].copy()
aqi_6m["date"].min(), aqi_6m["date"].max()
```

```python
area_day_counts = (aqi_6m.groupby(["state","area"])["date"].nunique().reset_index(name = "total_days"))
reliable_areas = area_day_counts[area_day_counts["total_days"] >= 90]
```

```python
aqi_6m_reliable = aqi_6m.merge(reliable_areas[["state", "area"]],on=["state", "area"], how="inner")
```

```python
area_summary = (aqi_6m_reliable.groupby(["state","area"]).agg(avg_aqi = ("aqi_value","mean"),
                                                              avg_severity = ("severity_rating","mean"), 
                                                              severe_days=("severity_rating", lambda x: (x >= 5).sum()),
                                                              total_days = ("severity_rating","count")).reset_index())
```

```python
area_summary["severe_day_percentage"] = (area_summary["severe_days"]/area_summary["total_days"])*100
```

#### **TOP AND BOTTOM POLLUTANTS BY REGION**

```python
aqi_pollutants = aqi[aqi["date"] >= "2022-01-01"].copy()
aqi_pollutants["pollutant_list"] = (aqi_pollutants["prominent_pollutants"].astype(str).str.split(","))
aqi_pollutants = aqi_pollutants.explode("pollutant_list")
aqi_pollutants["pollutant"] = aqi_pollutants["pollutant_list"].str.strip()
aqi_pollutants = aqi_pollutants[aqi_pollutants["pollutant"].notna()&(aqi_pollutants["pollutant"]!="")]
```

```python
regional_pollutant_count = (aqi_pollutants.groupby(["zone","pollutant"]).size().reset_index(name = "occurance_count"))
```

```python
def top_bottom_pollutants(df,n=2):
    top = df.nlargest(n,"occurance_count")
    bottom = df.nsmallest(n,"occurance_count")
    return pd.concat([top,bottom])
```

```python
regional_pollution_summary = (regional_pollutant_count.groupby("zone",group_keys = False).apply(top_bottom_pollutants))
regional_pollution_summary.head(10)
```
